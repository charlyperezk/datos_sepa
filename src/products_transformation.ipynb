productos_unidad_medida_presentacion units unify

```python
# Define a dictionary to map various unit representations to their standardized form
unit_mapping = {
    # Units of mass
    'gr': 'g', 'GR': 'g', 'Gr': 'g', 'Grm': 'g', 'GRM': 'g', 'grm': 'g', 'Gramos': 'g',
    'kg': 'kg', 'KG': 'kg', 'Kg': 'kg', 'KG ': 'kg', 'Kg ': 'kg', 'kgr': 'kg', 'kgm': 'kg', 'KGM': 'kg', 'Kilo': 'kg',
    
    # Units of volume
    'l': 'L', 'L': 'L', 'LT': 'L', 'lt': 'L', 'lt.': 'L', 'ltr': 'L', 'Litro': 'L',
    'ml': 'mL', 'ML': 'mL', 'Ml': 'mL', 'ml.': 'mL', 'Mililitro': 'mL',
    'cl': 'cL', 'CL': 'cL',
    'cc': 'cm³', 'CC': 'cm³', 'cc.': 'cm³', 'cm3': 'cm³', 'CM3': 'cm³', 'Cm3': 'cm³', 'Centimetro Cubico': 'cm³',
    'dm3': 'dm³',
    
    # Units of length
    'm': 'm', 'M': 'm', 'MT': 'm', 'mt.': 'm', 'mtr': 'm', 'Metros': 'm',
    'cm': 'cm', 'cmt': 'cm', 'centimetros': 'cm',
    
    # Units of area
    'm2': 'm²', 'M2': 'm²', 'm2 ': 'm²', 'cmq': 'cm²', 'dmq': 'dm²',
    
    # Count units
    'unidad': 'unidad', 'unida': 'unidad', 'UN': 'unidad', 'UNIDAD': 'unidad', 'un': 'unidad', 'un.': 'unidad',
    'Unit(s)': 'unidad', 'Unid': 'unidad', 'UNI': 'unidad', 'uni': 'unidad', 'Unidad': 'unidad', 'U': 'unidad',
    'Un': 'unidad', 'EA': 'unidad', 'UD': 'unidad',
    
    # Other units
    'paque': 'paquete', 'PC': 'paquete',
    'CS': 'caja', 'CJ': 'caja', 'CA': 'caja',
    'par': 'par',
    'Pie': 'pie',
    'hjs': 'hoja',
    'pl': 'pliego',
    'cu': 'cubo',
    
    # Special cases
    'Sin unidad de medida': 'Sin unidad de medida',
    'mi ': 'mi',  # Assuming 'mi' stands for 'milla' (mile)
}

# Function to standardize units
def standardize_unit(unit):
    unit = unit.strip()  # Remove leading/trailing whitespace
    return unit_mapping.get(unit, unit)  # Return mapped unit or original if not in mapping

# Apply the standardization to the column
df['productos_unidad_medida_presentacion'] = df['productos_unidad_medida_presentacion'].apply(standardize_unit)

# Display unique values after standardization
print(df['productos_unidad_medida_presentacion'].unique())

# Count of each standardized unit
unit_counts = df['productos_unidad_medida_presentacion'].value_counts()
print("\nCount of each standardized unit:")
print(unit_counts)
```
